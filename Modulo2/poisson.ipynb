<h2>Funci贸n de Probabilidad (PMF, por sus siglas en ingles) de Poisson</h2>

La funci贸n de probabilidad de Poisson describe la probabilidad de que un n煤mero espec铆fico de eventos ocurra en un intervalo de tiempo fijo, dado que estos eventos ocurren a una tasa constante promedio y de manera independiente entre s铆. Matem谩ticamente, esta funci贸n se expresa como: 

$$ P(k) = \frac{e^{-\lambda} \lambda^k}{k!} $$

Utilizando la ecuaci贸n de la funci贸n de probabilidad de Poisson, podemos generar gr谩ficas de barras que muestran la funci贸n de masa de probabilidad (PMF) para distintos casos. Para llevar a cabo esta tarea, primero debemos cargar las librer铆as correspondientes.


```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import math
import pandas as pd
```

Una vez cargadas las librer铆as, procederemos a definir los par谩metros de la distribuci贸n de Poisson para diferentes valores de  y rangos de k. 

```python
def poisson (mu, k_range):
    p_k = []
    for k in k_range:
        p = (math.exp((-1)*mu)*(mu**k))/math.factorial(k)
        p_k.append(p)

    return p_k
```

```python
def probaility_graph(prob):
    sns.lineplot(prob)
    plt.ylabel('Probabilidad estimado')
```

```python
mu = 1
k_range = range(1,15)
prob = poisson(mu, k_range)
probaility_graph(prob)
```

En un proceso de Poisson, podemos definir la unidad de tiempo como t. El proceso de Poisson puede modelarse con una tasa que depende de t. La  siguiente presenta la funci贸n de masa de probabilidad (PMF) para la ocurrencia de k eventos en un tiempo t.

$$ P(kt) = \frac{e^{-\lambda t} (\lambda t)^k}{k!} $$

Donde:
- P(kt) es la probabilidad de observar k eventos en un tiempo determinado
- 位 es la tasa promedio de ocurrencia de eventos por unidad de tiempo
- e es la base del logaritmo natural (aproximadamente 2.71828)
- K es el n煤mero de eventos observados
- t es el tiempo de dependencia  de la ocurrencia del evento k

```python
def poisson_time (mu, t, k_range):
    p_k = []
    for k in k_range:
        p = (math.exp((-1)*mu*t)*((mu*t)**k))/math.factorial(k)
        p_k.append(p)

    return p_k
```

```python
k_range = range(1,40)
mu = 5
for t in range(1, 5):
    prob = poisson_time(mu, t, k_range)
    probaility_graph(prob)
```

Por otro lado, la funci贸n de distribuci贸n acumulativa (CDF) se presenta

Esta funci贸n acumulativa representa la probabilidad de que el tiempo entre llegadas consecutivas sea menor o igual a un valor t.

Para simular los procesos de entre-llegada en Poisson de manera efectiva, se emplea el m茅todo de construcci贸n del inverso de la funci贸n de distribuci贸n acumulativa (CDF). Este enfoque se basa en la idea de que si tenemos una distribuci贸n exponencial con par谩metro 位 y su CDF esta definido. Para obtener tiempos de llegada aleatorios que sigan esta distribuci贸n, se utiliza el inverso de esta CDF, que se denota como:

$$ F^{-1}(t) = -\frac{\ln(1 - t)}{\lambda} $$

Al aplicar esta transformaci贸n, obtenemos tiempos de llegada t que siguen una distribuci贸n exponencial con par谩metro 位. Esto permite simular la ocurrencia de eventos en intervalos de tiempo de manera consistente con la distribuci贸n caracter铆stica de Poisson, donde los eventos ocurren de manera aleatoria pero con una tasa promedio 位.

Este m茅todo es esencial en la simulaci贸n de procesos de Poisson en diversas aplicaciones, como modelar llegadas de clientes a un servicio, ocurrencia de fallos en sistemas, o cualquier proceso donde los eventos ocurran de manera independiente y a una tasa constante promedio. La utilizaci贸n del inverso de la CDF garantiza que los tiempos de llegada simulados mantengan las propiedades estad铆sticas deseables de la distribuci贸n exponencial subyacente.





```python
def entre_llegada(mu, t_range):
    f_t = []
    for t in t_range:
        f = (-1)*math.log(1-t)/mu
        f_t.append(f)

    return f_t
```

En esta simulaci贸n vamos a generar los tiempos de entre-llegada en un proceso de Poisson, primero generamos 10 variables aleatorias uniformemente distribuidas entre 0 y 1. Cada una de estas variables representa el tiempo t en la distribuci贸n exponencial que modela el intervalo de tiempo entre eventos sucesivos. Luego, utilizamos el inverso de la funci贸n de distribuci贸n acumulativa de la distribuci贸n exponencial para determinar el tiempo de entre-llegada para cada una de estas variables.

- Paso a (Probabilidad): Representa la probabilidad de que ocurra exactamente X eventos durante el intervalo de tiempo considerado. 
- Paso b (Tiempo de entre-llegada): Utiliza el inverso de la funci贸n de distribuci贸n acumulativa de la distribuci贸n exponencial para obtener el tiempo entre-llegadas Xi correspondiente a cada evento i.

```python
time_prob = np.random.uniform(low=0, high=1, size = 10)
mu = 5
interarrival = entre_llegada(mu, time_prob)

df = pd.DataFrame()
df['time_prob'] = time_prob
df['interarrival'] = interarrival

print(df)
```
