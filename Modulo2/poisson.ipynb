<h1>Simulación de un Proceso de Poisson en Python</h1>

En el ámbito de la probabilidad y la estadística, el proceso de Poisson es un modelo fundamental para describir eventos que ocurren de manera aleatoria y a lo largo del tiempo. Este proceso es especialmente útil para modelar fenómenos en los que los eventos son independientes y ocurren a una tasa constante, como llamadas telefónicas en una central, llegadas de clientes a un servicio, o la emisión de partículas radioactivas.

Un proceso de Poisson se caracteriza por los siguientes principios:

- Independencia de los eventos: Los eventos ocurren de manera independiente entre sí.
- Tasa constante (λ): Los eventos ocurren a una tasa constante λ (lambda) por unidad de tiempo.
- No hay ocurrencia simultánea de eventos: En cualquier intervalo de tiempo infinitesimal, la probabilidad de que más de un evento ocurra es insignificante.

El objetivo de esta simulación es ilustrar cómo podemos generar y visualizar un proceso de Poisson utilizando el lenguaje de programación Python.

<h2>Función de Probabilidad (PMF, por sus siglas en ingles) de Poisson</h2>

La función de probabilidad de Poisson describe la probabilidad de que un número específico de eventos ocurra en un intervalo de tiempo fijo, dado que estos eventos ocurren a una tasa constante promedio y de manera independiente entre sí. Matemáticamente, esta función se expresa como: 

$$ P(k) = \frac{e^{-\lambda} \lambda^k}{k!} $$

Donde:
- P(k) es la probabilidad de observar k eventos
- λ es la tasa promedio de ocurrencia de eventos por unidad de tiempo
- e es la base del logaritmo natural (aproximadamente 2.71828)
- K es el número de eventos observados. 

Esta distribución es discreta, y se aplica en contextos donde los eventos ocurren de forma aleatoria pero con una frecuencia media conocida, como el número de llamadas telefónicas recibidas por minuto en una central telefónica o la cantidad de vehículos que pasan por un peaje en una hora.

<h2>Simulando en Python</h2>

Utilizando la ecuación de la función de probabilidad de Poisson, podemos generar gráficas de barras que muestran la función de masa de probabilidad (PMF) para distintos casos. La función de probabilidad de Poisson nos permite calcular la probabilidad de observar exactamente k eventos en un intervalo de tiempo dado, para una tasa de ocurrencia promedio λ.

Para llevar a cabo esta tarea, primero debemos cargar las librerías correspondientes.


```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import math
import pandas as pd
```

Una vez cargadas las librerías, procederemos a definir los parámetros de la distribución de Poisson para diferentes valores de 𝜆 y rangos de k. 

```python
def poisson (mu, k_range):
    p_k = []
    for k in k_range:
        p = (math.exp((-1)*mu)*(mu**k))/math.factorial(k)
        p_k.append(p)

    return p_k
```

```python
def probaility_graph(prob):
    sns.lineplot(prob)
    plt.ylabel('Probabilidad estimado')
```

```python
mu = 1
k_range = range(1,15)
prob = poisson(mu, k_range)
probaility_graph(prob)
```

En esta grafica podemos ver la probabilidad de observar exactamente k eventos en un intervalo de tiempo dado. El eje horizontal representa el número de eventos k, que en este caso varía de 1 a 15. Cada valor de k indica una posible cantidad de eventos que pueden ocurrir en el intervalo de tiempo considerado. El eje vertical muestra la probabilidad de que ocurra exactamente k eventos, calculada usando la función de masa de probabilidad de Poisson.

Con λ=1, esperamos que el número promedio de eventos sea 1 en el intervalo de tiempo.

Procedemos a generar las gráficas para diferentes valores de λ con el objetivo de observar cómo varía la distribución de Poisson en función de esta tasa de ocurrencia promedio. Al visualizar las distribuciones para λ=5, λ=10, λ=15, λ=20, λ=25, λ=30 y λ=35 en rangos específicos de k, podremos analizar el impacto de λ en la forma y la dispersión de la distribución. Esto nos permitirá entender mejor cómo la tasa promedio de eventos afecta la probabilidad de ocurrencia de diferentes números de eventos en un intervalo de tiempo dado.

```python
k_range = range(1, 45)
for mu in range(5,36,5):
    prob = poisson(mu, k_range)
    probaility_graph(prob)
```

En un proceso de Poisson, podemos definir la unidad de tiempo como t. El proceso de Poisson puede modelarse con una tasa que depende de t. La  siguiente presenta la función de masa de probabilidad (PMF) para la ocurrencia de k eventos en un tiempo t.

$$ P(kt) = \frac{e^{-\lambda t} (\lambda t)^k}{k!} $$

Donde:
- P(kt) es la probabilidad de observar k eventos en un tiempo determinado
- λ es la tasa promedio de ocurrencia de eventos por unidad de tiempo
- e es la base del logaritmo natural (aproximadamente 2.71828)
- K es el número de eventos observados
- t es el tiempo de dependencia  de la ocurrencia del evento k

```python
def poisson_time (mu, t, k_range):
    p_k = []
    for k in k_range:
        p = (math.exp((-1)*mu*t)*((mu*t)**k))/math.factorial(k)
        p_k.append(p)

    return p_k
```

Aquí podemos observar cómo cambia la distribución con un valor fijo de λ y variaciones en t.

```python
k_range = range(1,40)
mu = 5
for t in range(1, 5):
    prob = poisson_time(mu, t, k_range)
    probaility_graph(prob)
```

```python
k_range = range(1,65)
mu = 15
for t in range(1, 5):
    prob = poisson_time(mu, t, k_range)
    probaility_graph(prob)
```

Ahora procedemos a observar cómo varía la distribución con un valor específico de λ=15 y diferentes intervalos de tiempo. Este análisis nos permitirá entender cómo la tasa de ocurrencia más alta afecta la distribución de eventos en diferentes periodos de tiempo. Comparando las distribuciones para distintos valores de t, podemos visualizar cómo la probabilidad de observar diferentes cantidades de eventos cambia bajo una tasa de ocurrencia más alta y cómo se distribuyen estas probabilidades a lo largo del tiempo.

<h2> Proceso de Llegadas y Tiempo Entre Llegadas</h2>

Los procesos de entre-llegada son una aplicación fundamental de los procesos de Poisson. Mientras que el número de ocurrencias de eventos sigue una distribución discreta de Poisson, el intervalo de tiempo entre eventos consecutivos se modela con una distribución exponencial, que es continua. En este contexto, X_n representa el intervalo de tiempo entre llegadas sucesivas, donde X_1 es el tiempo desde el inicio del proceso hasta el primer evento, y X_4 sería el tiempo entre el tercer y el cuarto evento, y así sucesivamente. 

La función de densidad de probabilidad (PDF) para X_n se describe como

$$ f_{x_n}(t) = \lambda e^{-\lambda t} $$

, que en este caso es la función de densidad de probabilidad de la distribución exponencial donde:
- λ es la tasa promedio de ocurrencia de eventos por unidad de tiempo
- t≥0

Por otro lado, la función de distribución acumulativa (CDF) se presenta 

$$ F_{x_n}(t) = 1 - e^{-\lambda t} $$

Esta función acumulativa representa la probabilidad de que el tiempo entre llegadas consecutivas sea menor o igual a un valor t.

Estas ecuaciones son fundamentales para entender cómo los intervalos de tiempo entre eventos en un proceso de Poisson están distribuidos y cómo cambia la probabilidad con diferentes valores de λ.

Para simular los procesos de entre-llegada en Poisson de manera efectiva, se emplea el método de construcción del inverso de la función de distribución acumulativa (CDF). Este enfoque se basa en la idea de que si tenemos una distribución exponencial con parámetro λ y su CDF esta definido. Para obtener tiempos de llegada aleatorios que sigan esta distribución, se utiliza el inverso de esta CDF, que se denota como:

$$ F^{-1}(t) = -\frac{\ln(1 - t)}{\lambda} $$

Al aplicar esta transformación, obtenemos tiempos de llegada t que siguen una distribución exponencial con parámetro λ. Esto permite simular la ocurrencia de eventos en intervalos de tiempo de manera consistente con la distribución característica de Poisson, donde los eventos ocurren de manera aleatoria pero con una tasa promedio λ.

Este método es esencial en la simulación de procesos de Poisson en diversas aplicaciones, como modelar llegadas de clientes a un servicio, ocurrencia de fallos en sistemas, o cualquier proceso donde los eventos ocurran de manera independiente y a una tasa constante promedio. La utilización del inverso de la CDF garantiza que los tiempos de llegada simulados mantengan las propiedades estadísticas deseables de la distribución exponencial subyacente.
​




```python
def entre_llegada(mu, t_range):
    f_t = []
    for t in t_range:
        f = (-1)*math.log(1-t)/mu
        f_t.append(f)

    return f_t
```

En esta simulación vamos a generar los tiempos de entre-llegada en un proceso de Poisson, primero generamos 10 variables aleatorias uniformemente distribuidas entre 0 y 1. Cada una de estas variables representa el tiempo t en la distribución exponencial que modela el intervalo de tiempo entre eventos sucesivos. Luego, utilizamos el inverso de la función de distribución acumulativa de la distribución exponencial para determinar el tiempo de entre-llegada para cada una de estas variables.

- Paso a (Probabilidad): Representa la probabilidad de que ocurra exactamente X eventos durante el intervalo de tiempo considerado. 
- Paso b (Tiempo de entre-llegada): Utiliza el inverso de la función de distribución acumulativa de la distribución exponencial para obtener el tiempo entre-llegadas Xi correspondiente a cada evento i.

```python
time_prob = np.random.uniform(low=0, high=1, size = 10)
mu = 5
interarrival = entre_llegada(mu, time_prob)

df = pd.DataFrame()
df['time_prob'] = time_prob
df['interarrival'] = interarrival

print(df)
```

Para modelar el proceso de llegada en un proceso de Poisson, se utiliza la suma acumulativa de los tiempos de entre-llegada. Cada tiempo de llegada se denota como T1, T2, ..., Tn, donde Ti representa el momento en el que ocurre el i-ésimo evento. Estos tiempos de llegada están determinados por la suma acumulativa de los tiempos de entre-llegada X1, X2, ..., Xn.

Los tiempos de entre-llegada Xi siguen una distribución exponencial debido a la naturaleza del proceso de Poisson, donde la ocurrencia de eventos es aleatoria e independiente a una tasa promedio λ.

La relación entre los tiempos de entre-llegada y los tiempos de llegada se establece como sigue:
- T1 = X1:T1 es simplemente el tiempo de entre-llegada del primer evento.
- T2 = X1 + X2 es la suma del tiempo de entre-llegada del primer evento y el segundo evento.
- T3 = X1 + X2 + X3 es la suma acumulativa de los tiempos de entre-llegada del primer, segundo y tercer evento, y así sucesivamente hasta \( T_n \).

Estas sumas acumulativas siguen una distribución conocida como distribución Gamma, donde la suma de n variables exponenciales independientes con el mismo parámetro λ resulta en una variable con parámetros n y λ. 

Este enfoque permite modelar con precisión cómo se distribuyen los tiempos de llegada de eventos en un proceso de Poisson, capturando la naturaleza estocástica y la tasa promedio de ocurrencia de eventos.

De manera concisa y puntual, para obtener los tiempos de llegada en un proceso de Poisson, simplemente realizamos una suma acumulativa de los tiempos de entre-llegada. Cada tiempo de llegada Ti se calcula sumando secuencialmente los tiempos de entre-llegada X1, X2, ..., Xi.

```python
df['arrival'] = df['interarrival'].cumsum()
print(df)
```

Al combinar ambos enfoques, no solo podemos calcular la probabilidad de ocurrencia de eventos en un momento dado, sino también simular cómo se distribuyen estos eventos en el tiempo.
